```python
test_df.describe()
```

Similar case we see in test dataset

**Let's resolve the issue**

```python
## Property_size

train_df[train_df.property_size>8000]
```

It's not possible to have such a huge size with equivalent price that of others houses with lesser size.

So lets drop these rows.

```python
train_df = train_df[train_df.property_size<8000]
```

```python
## property_age

train_df[train_df.property_age>50]
```

Dropping these rows as this might not possible to have a house with more than 50 years age.

This may be possible but I am not going to consider it as 75% quartile (means 75% of houses in our dataset) is only 10.

```python
train_df = train_df[train_df.property_age<=50]
```

```python
## Bathroom

train_df[train_df.bathroom>6]
```

So it's clear from the above that there no such house exist with such a small size and lots of bathroom (20 bathrooms in 1100 property size).

```python
train_df = train_df[train_df.bathroom<=6]
```

```python
## cup_board

train_df[train_df.cup_board>=15]
```

It's not possible to have more than 15 cupboards.

There is one outlier as well with 48026 cup_borads in property_size of 700 sq.ft.

```python
train_df = train_df[train_df.cup_board<15]
```

```python
## Balconies

train_df = train_df[train_df.balconies<=5]
```

```python
## Floor

train_df[train_df.floor>18]
```

As 75% of houses are having max of 2 floors so I am going to intentionally drop these houses data having floor greater than 18 as it will affect my accuracy.

```python
train_df = train_df[train_df.floor<=18]
```

As there are very few houses with more than 4BHK and so lets make them 5

```python
# Now making a column 'bhk' which will be numeric and removing 'type' column

def find_bhk(s):
    if(s=='BHK4PLUS'):
        return 5
    return int(s[len(s)-1])

# print(type(find_bhk('RK4')))

train_df.type = train_df.type.apply(find_bhk)
test_df.type = test_df.type.apply(find_bhk)

train_df.rename(columns={'type':'bhk'},inplace=True)
```

## `Creating the Pipeline`

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score,mean_absolute_error
```

## `Importing Regression Models`

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

**Let's try different different Regression Models now ..**

## `Linear Regression`

```python
step1 = ColumnTransformer(transformers=[
    ('scal_tnf',MinMaxScaler(),[0,2,3,5,6,7,8,11,12,13,15,16,19]),
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[1,4,9,10,14,17,18])
],remainder='passthrough')

step2 = LinearRegression()

pipe=Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_val)
print('R2 score',r2_score(y_val,y_pred))
print('Mean Absolute error',mean_absolute_error(y_val,y_pred))
```

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor
```

## `Voting Regressor`

```python
step1 = ColumnTransformer(transformers=[
    ('scal_tnf',MinMaxScaler(),[0,2,3,5,6,7,8,11,12,13,15,16,19]),
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[1,4,9,10,14,17,18])
],remainder='passthrough')

rf = RandomForestRegressor(n_estimators=350,random_state=3,max_samples=0.5,bootstrap=True,max_features=0.75,max_depth=15)
gbdt = GradientBoostingRegressor(n_estimators=100,max_features=0.5)
xgb = XGBRegressor(n_estimators=25,learning_rate=0.3,max_depth=5)
et = ExtraTreesRegressor(n_estimators=100,random_state=3,max_samples=0.5,bootstrap=True,max_features=0.75,max_depth=10)

step2 = VotingRegressor([('rf', rf), ('gbdt', gbdt), ('xgb',xgb), ('et',et)],weights=[5,1,1,1])

pipe=Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_val)
print('R2 score',r2_score(y_val,y_pred))
print('Mean Absolute error',mean_absolute_error(y_val,y_pred))
```

## `Stacking Regressor`

```python
step1 = ColumnTransformer(transformers=[
    ('scal_tnf',MinMaxScaler(),[0,2,3,5,6,7,8,11,12,13,15,16,19]),
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[1,4,9,10,14,17,18])
],remainder='passthrough')

estimators = [
    ('rf', RandomForestRegressor(n_estimators=350,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)),
    ('gbdt',GradientBoostingRegressor(n_estimators=100,max_features=0.5)),
    ('xgb', XGBRegressor(n_estimators=25,learning_rate=0.3,max_depth=5))
]

step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

pipe=Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_val)
print('R2 score',r2_score(y_val,y_pred))
print('Mean Absolute error',mean_absolute_error(y_val,y_pred))
```

So Gradient Boosting Regressor is performing well and giving highest accuracy in case of Validation Accurcay i.e of 82.51%.
